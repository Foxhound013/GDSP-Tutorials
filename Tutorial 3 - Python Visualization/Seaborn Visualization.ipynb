# Data Visualization with Python
## Seaborn

Seaborn is the statistical offspring of Matplotlib. That is to say, it is built on top of Matplotlib and all of the rules of Matplotlib land still apply here. If you've been reading up on Matplotlib, no worries! You're hard earned knowledge is by no means a waste. Everything you learned still applied, you're just getting some easy and nice functionality on top of what is provided by Matplotlib to make your life as a scientist easier.

Note: If you happen to be familiar with lattice graphics panels (or even ggplot facets) in R, seaborn is probably the best way to replicate the functionality you might be used to in Python. We'll cover that breifly towards the end. Seaborn calls them facetgrids.

As with any good statistical software, seaborn has some datasets built in that we can call to get up and running!

```python
import matplotlib.pyplot as plt # It will become clear later why we're bringing matplotlib in alongside of seaborn.
import seaborn as sns
sns.set() # This will override the matplotlib styling, I tend to do this to make my plots prettier out of the box.
```

## The Data
We'll need a dataset to get going with any visualizations. Luckily seaborn has a few datasets built in that we can make use of.

```python
tips = sns.load_dataset('tips') # The dataset is a csv file, seaborn is leveraging pandas in the background to read in the data.
tips.head() # the .head function is a pandas method.
```

## Enter the Visualizations!

Let's start with two of the most basic plot we can, the scatterplot. You'll be using scatterplots for a variety of things in your career so this is as good a place as any to start.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips);
```

The cool thing about scatterplot is you can very easily color your points by categories in your dataset. Let's take an example from our data, what might be the difference in the distribution of the data for smokers and non-smokers?

(I recognize that this may not be the most sensical comparison but sometimes there are strange relationships that pop up in the data. It's worth slicing and dicing your data in as many ways as you can. We'll have more on this in a minute.)

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips);
```

There is a more general function that we can leverage here that provides a little more flexibility. It's called relplot. It will default to a scatterplot but gives the ability to make other types of plots like line plots.

```python
sns.relplot(x='total_bill', y='tip', data=tips);
```

While a line plot definitely isn't the right kind of graph for this data, we can do it with no protest from seaborn.

```python
sns.relplot(x='total_bill', y='tip', kind='line', data=tips);
```

## Scaterplot Matrices

If you're use to R, you may be missing scatter plot matrices. Seaborn brings that functionality into Python, making it much easier than trying to cook up your own solution in Matplotlib.

If you haven't worked with scatterplot matrices before, they're a great way of visualizing many possible relationships in your data all at once. There is a ton of information baked into a single plot to help get you off to a good start with your exploratory analysis.

```python
sns.pairplot(tips, hue='smoker')
```

## Faceting or Conditional Plots

To work with faceting or conditional plotting, we'll need to have matplotlib in our project explicitly as well. This is to tell python what type of plot to insert into the facets or panels.

These types of plots are extremely powerful to flesh out the relationship amongs different variables and can pack a ton of information into a single plot.

```python
# Below says to split the plots into columns based on category of Sex and color based on category of smoker.
g = sns.FacetGrid(tips, col="sex", hue="smoker") 
# below uses a scatter plot to represent the data, x as total bill and y as tip amount.
g.map(plt.scatter, "total_bill", "tip", alpha=.7) 
```

You'll notice the map method being used above. The map method is telling Python to map the plt.scatter method to g's FacetGrid object. The following arguments are telling plt.scatter to put the total bill on the x and tip on the y. The conditional information is all contained with the g's FacetGrid object. The columns should be sex and the coloration should be based on smoker or non-smoker.
