# Data Visualization with Python
## Seaborn

Seaborn is the statistical offspring of Matplotlib. That is to say, it is built on top of Matplotlib and all of the rules of Matplotlib land still apply here. What you learned in Matplotlib, yep, its still useful here!

Note: If you happen to be familiar with lattice graphics in R, seaborn is probably the best way to replicate that functionality in Python. We'll cover that breifly towards the end. Seaborn calls them facetgrids.

As with any good statistical software, seaborn has some datasets built in that we can call to get up and running!

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # This will override the matplotlib styling
# had we imported matplotlib as plt like before we could do what we did last time
# plt.style.use('seaborn')
```

```python
tips = sns.load_dataset('tips') # The dataset is a csv file, seaborn is leveraging pandas in the background to read in the data.
tips.head() # the .head function is from pandas.
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips);
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips);
```

There is a more general function that we can leverage here that provides a little more flexibility. It's called relplot. It will default to a scatterplot.

```python
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# While a line plot definitely isn't the right kind of graph for this data, we can do it with no protest from seaborn
ax = sns.relplot(x='total_bill', y='tip', kind='line', data=tips);
```

```python
# If you're use to R, you may be missing scatter plot matrices. Seaborn brings those pretty easily to Python.y
g = sns.PairGrid(tips)
g.map(plt.scatter)
```

```python
# Conditional plotting isn't too hard in seaborn either!
g = sns.FacetGrid(tips, col="sex", hue="smoker") # says to split the plots into columns based on category of Sex and color based on category of smoker.
g.map(plt.scatter, "total_bill", "tip", alpha=.7) # use a scatter plot to represent them, x as total bill and y as tip amount.
```
